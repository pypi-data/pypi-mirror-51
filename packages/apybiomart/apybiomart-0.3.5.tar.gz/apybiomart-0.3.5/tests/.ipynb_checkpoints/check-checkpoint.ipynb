```python
import apybiomart as apy
import pybiomart
import pandas as pd
import time
```

```python
%load_ext line_profiler
```

```python
aq = await apy.aquery(attributes=["refsnp_id", "chr_name", "chrom_start", "allele"], 
                filters={"chr_name": "MT", "start": "3308"}, 
                dataset="hsapiens_snp")
q = apy.query(attributes=["refsnp_id", "chr_name", "chrom_start", "allele"], 
              filters={"chr_name": "MT", "start": "3308"}, 
              dataset="hsapiens_snp")
da = pybiomart.Dataset(name='hsapiens_snp', host='http://www.ensembl.org')
p = da.query(attributes=['refsnp_id', "chr_name", "chrom_start", "allele"], 
                   filters={'chr_name': "MT", "start": "3308"})
```

```python
pd.testing.assert_frame_equal(q, aq)
pd.testing.assert_frame_equal(q, p)
pd.testing.assert_frame_equal(aq, p)
```
